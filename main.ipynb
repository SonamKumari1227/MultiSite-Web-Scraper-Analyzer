```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

def get_subjectivity_score(paragraph):
    # Initialize Sentiment Intensity Analyzer
    sid = SentimentIntensityAnalyzer()

    # Get sentiment scores
    sentiment_scores = sid.polarity_scores(paragraph)

    # Subjectivity score ranges from 0 to 1
    subjectivity_score = sentiment_scores['compound']

    return subjectivity_score

# Example usage
paragraph = "earthquake destroyed the life of many people. it was pathetic!"
subjectivity_score = get_subjectivity_score(paragraph)
print("Subjectivity Score:", subjectivity_score)
```

```python
def read_stop_words(file_path):
    with open(file_path, 'r') as file:
        stop_words = set(file.read().splitlines())
    return stop_words
```

```python
stop_words_files = ['./StopWords/StopWords_Auditor.txt', './StopWords/StopWords_Currencies.txt', './StopWords/StopWords_DatesandNumbers.txt', './StopWords/StopWords_Generic.txt', './StopWords/StopWords_GenericLong.txt', './StopWords/StopWords_Geographic.txt', './StopWords/StopWords_Names.txt']
combined_stop_words = set()

for file_path in stop_words_files:
    combined_stop_words.update(read_stop_words(file_path))
```

```python
def clean_paragraph(paragraph, combined_stop_words):
    # Split the paragraph into words
    words = paragraph.split()

    # Remove stop words from the list of words
    filtered_words = [word for word in words if word.lower() not in stop_words]

    # Join the filtered words back into a cleaned paragraph
    cleaned_paragraph = ' '.join(filtered_words)

    return cleaned_paragraph

```

```python
#average para length
import nltk

def average_sentence_length(paragraph):
    # Tokenize the paragraph into sentences
    sentences = nltk.sent_tokenize(paragraph)

    # Calculate the length of each sentence (number of words)
    sentence_lengths = [len(nltk.word_tokenize(sentence)) for sentence in sentences]

    # Find the average length of all sentences
    if sentence_lengths:
        average_length = sum(sentence_lengths) / len(sentence_lengths)
        return average_length
    else:
        return 0  # Handle the case where there are no sentences

# Example usage
paragraph = "This is a sample paragraph. It contains multiple sentences. Each sentence has a varying number of words."
avg_length = average_sentence_length(paragraph)
print("Average Sentence Length:", avg_length)
```

```python
import syllables

def is_complex(word, syllable_threshold=2):
    syllable_count = syllables.estimate(word)
    return syllable_count > syllable_threshold

def percentage_complex_words(paragraph, syllable_threshold=2):
    words = paragraph.split()
    complex_word_count = sum(is_complex(word, syllable_threshold) for word in words)
    total_word_count = len(words)

    if total_word_count == 0:
        return 0.0  # Avoid division by zero

    percentage_complex = (complex_word_count / total_word_count) * 100
    return round(percentage_complex, 2)

# Example usage
paragraph = "This is an example paragraph with some complex words."
percentage_complex = percentage_complex_words(paragraph)
print(f"Percentage of complex words: {percentage_complex}%")
```

```python
import syllables

def syllables_per_word(paragraph):
    # Split the paragraph into words
    words = paragraph.split()

    # Calculate the total number of syllables and words
    total_syllables = sum(syllables.estimate(word) for word in words)
    total_words = len(words)

    # Calculate the syllables per word
    syllables_per_word = total_syllables / total_words if total_words != 0 else 0

    return syllables_per_word

# Example usage:
paragraph = "This is an example paragraph with some words."
syllables_per_word_value = syllables_per_word(paragraph)
print("Syllables per Word:", syllables_per_word_value)
```

```python
def count_personal_pronouns(paragraph):
    # List of personal pronouns
    personal_pronouns = ['I', 'me', 'my', 'mine', 'myself', 'you', 'your', 'yours', 'yourself',
                         'he', 'him', 'his', 'himself', 'she', 'her', 'hers', 'herself',
                         'it', 'its', 'itself', 'we', 'us', 'our', 'ours', 'ourselves',
                         'you', 'your', 'yours', 'yourselves', 'they', 'them', 'their', 'theirs', 'themselves']

    # Count the occurrences of personal pronouns
    count = sum(paragraph.lower().count(pronoun) for pronoun in personal_pronouns)

    return count

# Example usage:
paragraph = "I like to watch movies. You should join us."
pronoun_count = count_personal_pronouns(paragraph)
print("Personal Pronoun Count:", pronoun_count)
```

```python
from textstat import syllable_count

def fog_index(paragraph):
    words = paragraph.split()
    total_words = len(words)
    total_sentences = paragraph.count('.') + paragraph.count('!') + paragraph.count('?')

    if total_sentences == 0:
        return 0

    avg_sentence_length = total_words / total_sentences
    complex_words = [word for word in words if syllable_count(word) > 2]
    percentage_complex_words = (len(complex_words) / total_words) * 100

    fog_index = 0.4 * (avg_sentence_length + percentage_complex_words)
    return fog_index

# Example usage:
paragraph = "This is a sample paragraph with various words of different lengths."
fog_score = fog_index(paragraph)
print("FOG Index:", fog_score)
```

```python
import re
def avg_words_per_sentence(paragraph):
    sentences = [sentence.strip() for sentence in re.split(r'[.!?]', paragraph) if sentence.strip()]
    
    if not sentences:
        return 0

    total_words = sum(len(sentence.split()) for sentence in sentences)
    avg_words_per_sentence = total_words / len(sentences)
    
    return avg_words_per_sentence

# Example usage:
paragraph = "This is a sample paragraph. It has multiple sentences! Each sentence varies in length."
average_words = avg_words_per_sentence(paragraph)
print("Average Number of Words per Sentence:", average_words)
```
